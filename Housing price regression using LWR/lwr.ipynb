```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
data=pd.read_csv('Housing_Price_data_set.csv')
data['num_driveway']=data.driveway.map({'yes':1,'no':0})
data['num_recroom']=data.recroom.map({'yes':1,'no':0})
data['num_fullbase']=data.fullbase.map({'yes':1,'no':0})
data['num_gashw']=data.gashw.map({'yes':1,'no':0})
data['num_airco']=data.airco.map({'yes':1,'no':0})
data['num_prefarea']=data.prefarea.map({'yes':1,'no':0})
X=data[['lotsize','bedrooms','bathrms','stories','num_driveway','num_recroom','num_fullbase','num_gashw','num_airco','garagepl','num_prefarea']].values
Y=data[['price']].values
print(X[0])
```

```python
ones = np.ones([X.shape[0],1])
original_X = np.concatenate((ones,X),axis=1)
X=(original_X-np.mean(original_X))/np.std(original_X)
Y=(Y-np.mean(Y))/np.std(Y)

x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=50,test_size=.3)
print(x_train)

theta = np.transpose(np.matrix(np.zeros(12)))

#set gradient parameters
alpha = 0.001
iters = 10000
```

```python
##calculating locally weighted regression ##
theta_0=0
def lwr(X,Y,weight,theta_0,theta,iters,alpha):
    cost=[]
    xTrans=X.transpose()
    for i in range(0,iters):
        hypothesis=X*theta
        loss=hypothesis-Y
        cost.append(np.sum(weight.T*np.power(loss,2))/(2*len(X)))
        if i>0:
            print("iteration=",i,"| cost=",cost[i])
        gradient=(weight.T*loss)/len(X)
        theta=theta-(alpha*(gradient))
    return theta,cost

i=85 ##suppose we want to predict the price of 85th point 
xdataset=[]
ydataset=[]
count=0
for j in range(0,len(x_train)):
    if(np.array_equal(x_train[i],x_train[j])!=True and np.absolute(x_train[j][1]-x_train[i][1])<=4):
        xdataset.append(x_train[j])
        ydataset.append(y_train[j])
        count=count+1
    if(count>=100):
        break
xdataset=np.array(xdataset)
ydataset=np.array(ydataset)
givenpoint=np.array(x_train[i]) #point whose prediction is to be done
print(xdataset.shape)
weight=[]
for i in range(0,len(xdataset)):
    temp=xdataset[i][1]-givenpoint[1]
    temp2=(-1)*np.power(temp,2)/(2*0.99*0.99)
    weight.append(np.exp(temp2))
    

weight=np.array(weight)
print(weight.shape)
g,cost=lwr(xdataset,ydataset,weight,theta_0,theta,iters,alpha)
print("parameters after lwr for prediction of 100th point:",g);
y_pred=np.matmul(x_train[85],g)
print("calculated value for given point is",y_pred)
print("The original value for given point is",y_train[85])
            
```
